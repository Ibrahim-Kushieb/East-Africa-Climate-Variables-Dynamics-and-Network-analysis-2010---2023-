# **Temperature Time Series**

```python
import pandas as pd
import matplotlib.pyplot as plt



# Assuming your data columns are organized like this:
# 'Date' | 'Column1' | 'Column2' | ... | 'ColumnN'
#df = df.drop(columns=['Region'])
# Convert the 'Date' column to datetime if it's not already
T['Date'] = pd.to_datetime(T['Date'])

# Set 'Date' as the index for time series plotting
T.set_index('Date', inplace=True)

# Get a list of all column names (excluding 'Date')
columns_to_plot = T.columns[T.columns != 'Date']

# Create time series plots for all columns
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

offset = 0.001  # Offset value to separate the two identical columns visually

for i, column in enumerate(columns_to_plot):
    if i % 2 == 0:
        offset *= -1  # Alternate the offset direction for each column
    plt.plot(T.index, T[column] + offset, marker='o', linestyle='-', markersize=5, label=column)

# Customize the plot
plt.title('Temperature Time Series for EA (2010 - 2023)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
